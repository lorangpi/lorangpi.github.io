```python
# Load a mp4 vide oand compress it to a smaller size using ffmpeg
import subprocess
import os
import sys
def compress_video(input_file, output_file, crf=23):
    command = [
        'ffmpeg',
        '-i', input_file,
        '-vcodec', 'libx264',
        '-crf', str(crf),
        output_file
    ]
    subprocess.run(command, check=True)

# Get all the mp4 files in the current directory
def get_mp4_files():
    mp4_files = []
    print(os.listdir('.'))
    for file in os.listdir('.'):
        if file.endswith('.mp4'):
            mp4_files.append(file)
        if file.endswith('.MOV'):
            mp4_files.append(file)
    return mp4_files
print(get_mp4_files())
for mp4_file in get_mp4_files():
    output_file = f'compressed_{mp4_file}'
    print(f'Compressing {mp4_file} to {output_file}...')
    compress_video(mp4_file, output_file, crf=35)
    print(f'Finished compressing {mp4_file}.')
```

```python
# For file in get_mp4_files(), replace the original file with the compressed file if the compressed file is smaller than the original file
for mp4_file in get_mp4_files():
    compressed_file = f'compressed_{mp4_file}'
    if os.path.exists(compressed_file):
        original_size = os.path.getsize(mp4_file)
        compressed_size = os.path.getsize(compressed_file)
        if compressed_size < original_size:
            os.replace(compressed_file, mp4_file)
            print(f'Replaced {mp4_file} with {compressed_file} because it is smaller.')
        else:
            os.remove(compressed_file)
            print(f'Removed {compressed_file} because it is not smaller than {mp4_file}.')
```

```python
video_path = "day_iphone_forklift.MOV"

# Compress the video and replace if the compressed file is smaller
output_file = "compressed_day_iphone_forklift.mp4"
compress_video(video_path, output_file, crf=30)
if os.path.exists(output_file):
    original_size = os.path.getsize(video_path)
    compressed_size = os.path.getsize(output_file)
    if compressed_size < original_size:
        os.replace(output_file, video_path)
        print(f'Replaced {video_path} with {output_file} because it is smaller.')
    else:
        os.remove(output_file)
        print(f'Removed {output_file} because it is not smaller than {video_path}.')
```
